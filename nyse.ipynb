```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web

style.use('ggplot')
```

```python
start = dt.datetime(2000,1,1)
end = dt.datetime(2019,12,31)

df = web.DataReader('GOOGL','yahoo',start,end)
```

```python
print(df)
```

```python
plt.plot(df)
plt.show()
```

```python
df['Adj Close'].plot()
```

```python
#plotting a 100 moving average
"""
Takes today's stock prices along with the 99 prior day's prices
and averages them   


""" 
df['100ma'] = df['Adj Close'].rolling(window=100).mean()
```

```python
print(df.head())
"""
Note the NaN since the head() only plots the first few rows, so there
isn't 100 rows too plot.

However, should we use df.tail(), the 100 day moving average makes more
sense!
"""
```
